```python
def CosineSim(list1, list2):
    dot = 0
    """ Calculates cosine similairty bewteen two vectors """
    for k in range(0,len(A)):
        dot += list1[k]*list2[k]
    normA = np.linalg.norm(list1)
    normB = np.linalg.norm(list2)
    return dot/(normA*normB) # sim

# Make a dict: key - user, value - list of cosine similairty of matches
friendCos = {}
for key in matchesFriend.keys():
    cosSims = []
    A = list(cleaned.iloc[key-1].values)
    for friend in matchesFriend[key]:
        B = list(cleaned.iloc[int(friend)-1].values)
        cosSims.append(CosineSim(A,B))
    friendCos[key] = cosSims


FCavg = {}
# Get the average of cosine similiarty of your friends
for key in friendCos.keys():
    FCavg[key] = sum(friendCos[key])/len(friendCos[key])
model_df = cleaned
model_df["FCavg"] = pd.Series(FCavg)
```

```python
# Two-step approach
def Find_Friends(person):
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.tree import DecisionTreeClassifier
    
    cosineAll = {}
    diff = {}
    for name in model_df.index:
        cosineAll[name] = CosineSim(list(model_df.iloc[int(person) - 1].values),list(model_df.iloc[int(name) - 1].values))
        diff[name] = FCavg[person] - cosineAll[name]
    def preproc():
        """ Create a dataframe for a specific user"""
        model_df1 = model_df
        model_df1["Friend?"] = pd.Series({value : 1 for value in matchesFriend[person]})
        model_df1["Cos"] = pd.Series(cosineAll)
        model_df1["Diff"] = pd.Series(diff)
        
        model_df1 = model_df1.fillna(0)  #model_df1 has 1 for positive labels 0 for unlabeled
        model_df1 = model_df1.drop(person) # removes person from data, looks at every one else
        return model_df1
    
    df = preproc()
    y = df["Friend?"]
    ys = 2*y - 1                    # pos = 1, unlabled = -1
    X = df.drop("Friend?", axis = 1)
    
    def two_step(X,y,ys):
        rf = RandomForestClassifier(n_estimators = 1000)
        rf.fit(X, y)
    
        pred = rf.predict_proba(X)[:,1]
        range_P = [min(pred * (ys > 0)), max(pred * (ys > 0))] # Min and Max values for known positives
    
        iP_new = ys[(ys < 0) & (pred >= range_P[1])].index
        iN_new = ys[(ys < 0) & (pred <= range_P[0])].index
        ys.loc[iP_new] = 1
        ys.loc[iN_new] = 0
    
        rf2 = RandomForestClassifier(n_estimators = 1000)
        
        for i in range(10):
            # If step 1 didn't find new labels, end
            if len(iP_new) + len(iN_new) == 0 and i > 0:
                break
            
            # Step 2
            # Retrain on new labels and get new scores
            rf2.fit(X, ys)
            
            pred = rf2.predict_proba(X)[:,-1]
            range_P = [min(pred * (ys > 0)), max(pred * (ys > 0))]

        
            # Repeat step 1
            iP_new = ys[(ys < 0) & (pred >= range_P[1])].index
            iN_new = ys[(ys < 0) & (pred <= range_P[0])].index
            ys.loc[iP_new] = 1
            ys.loc[iN_new] = 0
            
        return ys,pred
        
    return two_step(X,y,ys)

# Creates a dict: key - user, value - all possible friends (positive and unlabeled values)
posb_Fri = {}
for p in range(1,161):
    try:
        Labels,preds = Find_Friends(p)
        posb_Fri[p] = list(Labels[(Labels != 0)].index)
    except:
        pass
    
```
