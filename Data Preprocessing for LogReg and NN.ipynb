# Creating positive dataset

```python
#finding all the matches regardless of friend or romantic relation in the survey data, each person who states a match will count

Data = df
matchesFriend = {}
romantic = {}
i = 0

while i < len(Data):
    Friends = []
    romance = []
#     for j in range(16,22):
#         if Data.iloc[i][j] > 0:
#             romance.append(Data.iloc[i][j])
    for k in range(2,22):
        if Data.iloc[i][k] > 0:
            Friends.append(Data.iloc[i][k])
            matchesFriend[Data.iloc[i][0]] = Friends
    if romance != []:
        romantic[Data.iloc[i][0]] = romance
    i += 1
```

```python
# creating empty dataframe for calculation between matches for positive dataset
rows = []

ids=[]
for key in matchesFriend.keys():
    for friend in matchesFriend[key]:
        row = (df1.iloc[int(key)-1].values) - (df1.iloc[int(friend)-1].values)
        id1 = str(key)+ '&' + str(friend)
        ids.append(id1)        
        rows.append(row)

df_obs = pd.DataFrame(rows, columns=df1.columns).abs()
df_obs = df_obs.apply(lambda x: x.fillna(x.mean()),axis=0)
df_ids = pd.DataFrame(ids)
df_ids.columns=['U_ID']
df_obs_1 = pd.concat([df_ids,df_obs],axis=1)
df_obs_1['U_ID'] = df_obs_1['U_ID'].str.split('.').str[0]
df_obs_1.drop(columns=['ID'],inplace=True)
df_obs_1['y'] = 1
df_obs_1.head()
```

# Creating unlabeled dataset

```python
#use combinations to find all possible combinations of people

import itertools 
ul_set = list(itertools.combinations(list(range(1,161)),2))
```

```python
#create dictionary to convert list of pairs into dictionary

ul_dicts = {}
for elem in ul_set:
    try:
        ul_dicts[elem[1]].append(elem[0])
    except KeyError:
        ul_dicts[elem[1]] = [elem[0],]
```

```python
rows=[]
ids2=[]
for key in ul_dicts.keys():
    for friend in ul_dicts[key]:
        row = (df1.iloc[int(key)-1].values) - (df1.iloc[int(friend)-1].values)
        id2 = str(key)+ ' & ' + str(friend)
        ids2.append(id2)
        #print(row)         
        rows.append(row)
        #print(key)
        #print(friend)
        #print(str(key)+ ' ' + str(friend))

#create dataframe with the newly created row
df_obs2 = pd.DataFrame(rows, columns=df1.columns).abs() #take the absolute values of the differences
df_obs2 = df_obs2.apply(lambda x: x.fillna(x.mean()),axis=0) #fill any NaNs with the mean of the column
df_ids2 = pd.DataFrame(ids2) #create a dataframe for the unique ids
df_ids2.columns=['U_ID'] #rename new columbn
df_obs2_2 = pd.concat([df_ids2,df_obs2],axis=1) #combine the dataframes
df_obs2_2['y'] = -1 #add the label for unlabeled
df_obs2_2.drop(columns=['ID'],inplace=True) #drop ID column because no longer useful
df_obs2_2.head()
```

# Creating Positive and Unlabeled Dataset

```python
#combining the two datasets

dataset_df = pd.concat([df_obs_1, df_obs2_2], ignore_index=True)
dataset_df.head()
```
